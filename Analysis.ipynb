# Read data

## Weather

```python
weather_df = pd.read_csv('weather.csv',low_memory=False,index_col=0)
weather_df.index = pd.to_datetime(weather_df.index,unit='D')
weather_df['PrecipitationIn'] = weather_df['PrecipitationIn'].replace('T',np.nan)
weather_df['PrecipitationIn'] = weather_df['PrecipitationIn'].dropna().astype(float)
weather_df.tail()
```

```python
weather_df.columns
```

## Assault

```python
assault_df = pd.read_csv('assault.csv',low_memory=False)
assault_df['Datetime'] = pd.to_datetime(assault_df['Date'],format="%m/%d/%Y %I:%M:%S %p")
assault_df['Date'] = assault_df['Datetime'].apply(lambda x:x.date())
assault_df.head()
```

```python
assault_df_gb = assault_df.groupby('Date')['ID'].agg(len)
assault_df_gb.index = pd.to_datetime(assault_df_gb.index,unit='D')
```

## Murder

```python
murder_df = pd.read_csv('murder.csv',low_memory=False)
murder_df['Datetime'] = pd.to_datetime(murder_df['Date'],format="%m/%d/%Y %I:%M:%S %p")
murder_df['Date'] = murder_df['Datetime'].apply(lambda x:x.date())
murder_df.head()
```

```python
murder_df_gb = murder_df.groupby('Date')['ID'].agg(len)
murder_df_gb.index = pd.to_datetime(murder_df_gb.index,unit='D')
```

# Join data and cleanup

```python
ts_df = pd.DataFrame({'Gun_Assaults':assault_df_gb,
                      'Murders':murder_df_gb,
                      'Temperature':weather_df['Mean TemperatureF'],
                      'Precipitation':weather_df['PrecipitationIn'],
                      'Humidity':weather_df[' Mean Humidity']})

ts_df['Time'] = range((max(ts_df.index)-min(ts_df.index)).days+1)

s = ts_df.reset_index()['index'].apply(lambda x:x.week)
s.index = ts_df.index
ts_df['Week'] = s
```

# Power spectrum

```python
d = ts_df['Temperature'].resample('M').reset_index()['Temperature'].dropna().values
ps = np.fft.fft(d)**2
plt.plot(np.abs(ps))
plt.yscale('log')
#plt.xscale('log')
```

```python
n = len(d)
freq = np.fft.fftfreq(n,86400.)

d_shifted = np.fft.fftshift(d)
freq_shifted = np.fft.fftshift(freq)
plt.plot(freq_shifted,np.abs(d_shifted))
```

```python
half_n = np.ceil(n/2.0)
fft_x_half = (2.0 / n) * d[:half_n]
freq_half = freq[:half_n]
plt.plot(freq_half, np.abs(fft_x_half))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
```
